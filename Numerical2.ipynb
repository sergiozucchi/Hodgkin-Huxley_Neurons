```python
import matplotlib.pyplot as plt
from HH_model import HH_system

#Ignores RunTimeWarnings
import warnings
warnings.filterwarnings('ignore')
```

```python
exc2=HH_system(neurons={"1":{"V0":-7,"n0":0.52,"m0":0.02,"h0":0.31,"isExcitatory":True},
                        "2":{"V0":5,"n0":0.39,"m0":0.09,"h0":0.46,"isExcitatory":True}},
              r0=[0.9,0.1],
              synp=[
                  {"pre":"1","post":"2" ,"g":10},
                  {"pre":"2","post":"1" ,"g":10}
              ])
sol_exc=exc2.run_sys(100,0.05)
```

```python
t=sol_exc.t
V1=sol_exc.y[0]
V2=sol_exc.y[4]

plt.plot(t,V1,label="Excitatory 1")
plt.plot(t,V2,"--",label="Excitatory 2")
plt.legend(loc="upper right")
plt.xlabel("t [ms]")
plt.ylabel("V [mV]")
#plt.savefig("Exc2.pdf",bbox_inches="tight")
plt.show()
```

```python
sys3_g20=HH_system(neurons={"M":{"V0":-7,"n0":0.52,"m0":0.02,"h0":0.31,"isExcitatory":True},
                            "S":{"V0":5,"n0":0.39,"m0":0.09,"h0":0.46,"isExcitatory":True},
                            "I":{"V0":4,"n0":0.735,"m0":0.453,"h0":0.076,"isExcitatory":False}},
              r0=[0.9,0.1,0.5],
              synp=[
                  {"pre":"M","post":"S" ,"g":10},
                  {"pre":"S","post":"I" ,"g":10},
                  {"pre":"I","post":"S" ,"g":20}
              ])
sol_sysg20=sys3_g20.run_sys(100,0.05)
```
